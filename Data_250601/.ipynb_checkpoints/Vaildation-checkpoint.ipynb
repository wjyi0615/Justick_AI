```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from einops import rearrange

# 1. 데이터 불러오기 및 전처리
df = pd.read_csv("store/cabbage_separated.csv")
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
df = df.sort_values('date')

# 예측에 사용할 feature 선택
features = ['intake', 'gap', 'avg_price']
df = df[features + ['date']].dropna().reset_index(drop=True)

# 스케일링
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[features])

# 2. DLinear 모델 정의
class DLinear(nn.Module):
    def __init__(self, seq_len, pred_len, input_dim):
        super(DLinear, self).__init__()
        self.seq_len = seq_len
        self.pred_len = pred_len
        self.input_dim = input_dim
        self.linear = nn.Linear(seq_len, pred_len)

    def forward(self, x):
        # x: [B, T, C] → [B, C, T]
        x = rearrange(x, 'b t c -> b c t')
        x = self.linear(x)
        # 다시 [B, C, pred_len] → [B, pred_len, C]
        out = rearrange(x, 'b c t -> b t c')
        return out

# 3. 데이터셋 클래스 정의
class TimeSeriesDataset(Dataset):
    def __init__(self, data, seq_len, pred_len):
        self.data = data
        self.seq_len = seq_len
        self.pred_len = pred_len

    def __len__(self):
        return len(self.data) - self.seq_len - self.pred_len

    def __getitem__(self, idx):
        seq_x = self.data[idx:idx + self.seq_len]
        seq_y = self.data[idx + self.seq_len : idx + self.seq_len + self.pred_len]
        return torch.tensor(seq_x, dtype=torch.float32), torch.tensor(seq_y, dtype=torch.float32)

# 4. 학습 및 예측 함수
def train_dlinear(model, dataloader, epochs=30, lr=0.001):
    model.train()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.MSELoss()
    for epoch in range(epochs):
        for x, y in dataloader:
            optimizer.zero_grad()
            output = model(x)
            loss = criterion(output, y)
            loss.backward()
            optimizer.step()

def forecast(model, data, seq_len):
    model.eval()
    with torch.no_grad():
        input_seq = torch.tensor(data[-seq_len:], dtype=torch.float32).unsqueeze(0)  # [1, seq_len, C]
        pred = model(input_seq)  # [1, pred_len, C]
        return pred.squeeze(0).numpy()

# 5. 예측 수행
def run_prediction(pred_len):
    seq_len = 30
    dataset = TimeSeriesDataset(scaled_data, seq_len, pred_len)
    dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
    model = DLinear(seq_len, pred_len, input_dim=3)
    train_dlinear(model, dataloader)

    pred_scaled = forecast(model, scaled_data, seq_len)
    pred = scaler.inverse_transform(np.concatenate([np.zeros_like(pred_scaled), pred_scaled], axis=0))[:, -1]
    return pred

# 6. 예측 실행 및 시각화
pred_1 = run_prediction(1)
pred_7 = run_prediction(7)
pred_28 = run_prediction(28)

print("1일 예측 평균가격:", pred_1)
print("7일 예측 평균가격:", pred_7)
print("28일 예측 평균가격:", pred_28)

# 2025년 5월 이후 실제값
after_may_df = df[df['date'] >= '2025-05-01']

# 예측 날짜 생성 (마지막 날짜 이후 28일)
last_date = df['date'].iloc[-1]
future_dates = [last_date + pd.Timedelta(days=i) for i in range(1, 29)]

# 예측 결과 중 avg_price만 추출
pred_28_prices = pred_28[-28:]  # 만약 pred_28 shape이 [28]이라면 이 줄은 그대로, 아니라면 슬라이싱 수정 필요

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(after_may_df['date'], after_may_df['avg_price'], label='실제 평균가격')
plt.plot(future_dates, pred_28_prices, 'ro-', label='28일간 예측 평균가격')

plt.title("DLinear 배추 평균가격 예측 (향후 28일간)")
plt.xlabel("날짜")
plt.ylabel("가격 (원)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from einops import rearrange
import matplotlib.pyplot as plt

# 1. DLinear 모델 정의
class DLinear(nn.Module):
    def __init__(self, seq_len, pred_len, input_dim):
        super(DLinear, self).__init__()
        self.linear = nn.Linear(seq_len, pred_len)
        self.seq_len = seq_len
        self.pred_len = pred_len

    def forward(self, x):
        x = rearrange(x, 'b t c -> b c t')       # [B, C, T]
        out = self.linear(x)                     # [B, C, pred_len]
        out = rearrange(out, 'b c t -> b t c')   # [B, pred_len, C]
        return out

# 2. Dataset 클래스
class TimeSeriesDataset(Dataset):
    def __init__(self, data, seq_len, pred_len):
        self.data = data
        self.seq_len = seq_len
        self.pred_len = pred_len

    def __len__(self):
        return len(self.data) - self.seq_len - self.pred_len

    def __getitem__(self, idx):
        x = self.data[idx:idx+self.seq_len]
        y = self.data[idx+self.seq_len:idx+self.seq_len+self.pred_len]
        return torch.tensor(x, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

# 3. 데이터 로드
df = pd.read_csv("store/cabbage_separated.csv")
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
df = df.sort_values('date')

features = ['intake', 'gap', 'avg_price']
results = []

# 4. 연도별 예측 루프
for year in [2022, 2023, 2024, 2025]:
    train_start = f'{year-1}-06-01'
    train_end   = f'{year}-04-30'
    test_start  = f'{year}-05-01'
    test_end    = f'{year}-05-28'

    df_train = df[(df['date'] >= train_start) & (df['date'] <= train_end)].copy()
    df_test  = df[(df['date'] >= test_start) & (df['date'] <= test_end)].copy()

    if len(df_test) < 28 or len(df_train) < 60:
        print(f"{year} 데이터 부족 - 스킵")
        continue

    scaler = StandardScaler()
    train_scaled = scaler.fit_transform(df_train[features])
    test_scaled = scaler.transform(df_test[features])

    dataset = TimeSeriesDataset(train_scaled, seq_len=60, pred_len=28)
    dataloader = DataLoader(dataset, batch_size=16, shuffle=True)

    model = DLinear(seq_len=60, pred_len=28, input_dim=3)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
    criterion = nn.MSELoss()

    model.train()
    for epoch in range(20):
        for x, y in dataloader:
            optimizer.zero_grad()
            output = model(x)
            loss = criterion(output, y)
            loss.backward()
            optimizer.step()

    # 5. 마지막 구간으로부터 예측
    last_input = torch.tensor(train_scaled[-60:], dtype=torch.float32).unsqueeze(0)
    model.eval()
    with torch.no_grad():
        pred_scaled = model(last_input).squeeze(0).numpy()
        pred = scaler.inverse_transform(pred_scaled)

    # 6. 성능 측정
    gt = df_test[features].values[:28]
    y_true = gt[:, -1]
    y_pred = pred[:, -1]

    rmse = mean_squared_error(y_true, y_pred, squared=False)
    mae = mean_absolute_error(y_true, y_pred)

    results.append((year, rmse, mae))
    print(f"[{year}] RMSE: {rmse:.2f} / MAE: {mae:.2f}")

    # 7. 시각화
    plt.figure(figsize=(10, 4))
    plt.plot(df_test['date'][:28], y_true, label='실제 평균가격')
    plt.plot(df_test['date'][:28], y_pred, label='예측 평균가격')
    plt.title(f'{year}년 5월 평균가격 예측 (DLinear)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# 8. 전체 결과 정리
print("\n전체 연도별 성능:")
for y, r, m in results:
    print(f" - {y}년 5월 → RMSE: {r:.2f}, MAE: {m:.2f}")
```

```python
class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, num_layers=2, output_dim=3, pred_len=28):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)
        self.linear = nn.Linear(hidden_dim, output_dim * pred_len)
        self.pred_len = pred_len
        self.output_dim = output_dim

    def forward(self, x):
        out, _ = self.lstm(x)  # [B, T, H]
        out = out[:, -1, :]    # [B, H]
        out = self.linear(out) # [B, 28 * output_dim]
        out = out.view(-1, self.pred_len, self.output_dim)  # [B, 28, 3]
        return out
```

```python
from xgboost import XGBRegressor

def prepare_xgb_dataset(data, seq_len=60, pred_len=28):
    X, y = [], []
    for i in range(len(data) - seq_len - pred_len):
        X.append(data[i:i+seq_len, :].flatten())     # [60, 3] → [180]
        y.append(data[i+seq_len:i+seq_len+pred_len, -1])  # avg_price만 예측
    return np.array(X), np.array(y)
```

```python
# 공통 설정
seq_len = 60
pred_len = 28
features = ['intake', 'gap', 'avg_price']
results_lstm = []
results_xgb = []

for year in [2022, 2023, 2024, 2025]:
    train_start = f'{year-1}-06-01'
    train_end   = f'{year}-04-30'
    test_start  = f'{year}-05-01'
    test_end    = f'{year}-05-28'

    df_train = df[(df['date'] >= train_start) & (df['date'] <= train_end)].copy()
    df_test  = df[(df['date'] >= test_start) & (df['date'] <= test_end)].copy()

    if len(df_test) < 28 or len(df_train) < 60:
        print(f"{year} 데이터 부족 - 스킵")
        continue

    scaler = StandardScaler()
    train_scaled = scaler.fit_transform(df_train[features])
    test_scaled = scaler.transform(df_test[features])

    # ---------- LSTM ----------
    dataset = TimeSeriesDataset(train_scaled, seq_len, pred_len)
    dataloader = DataLoader(dataset, batch_size=16, shuffle=True)

    lstm_model = LSTMModel(input_dim=3, pred_len=pred_len)
    optimizer = torch.optim.Adam(lstm_model.parameters(), lr=1e-3)
    criterion = nn.MSELoss()

    lstm_model.train()
    for epoch in range(20):
        for x, y in dataloader:
            optimizer.zero_grad()
            output = lstm_model(x)
            loss = criterion(output, y)
            loss.backward()
            optimizer.step()

    lstm_model.eval()
    with torch.no_grad():
        last_input = torch.tensor(train_scaled[-seq_len:], dtype=torch.float32).unsqueeze(0)
        pred_lstm_scaled = lstm_model(last_input).squeeze(0).numpy()
        pred_lstm = scaler.inverse_transform(pred_lstm_scaled)

    gt = df_test[features].values[:pred_len]
    y_true = gt[:, -1]
    y_pred = pred_lstm[:, -1]
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    mae = mean_absolute_error(y_true, y_pred)
    results_lstm.append((year, rmse, mae))

    # ---------- XGBoost ----------
    X_train, y_train = prepare_xgb_dataset(train_scaled, seq_len, pred_len)
    xgb_model = XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.1)
    xgb_model.fit(X_train, y_train)

    last_x = train_scaled[-seq_len:, :].flatten().reshape(1, -1)
    pred_xgb_scaled = xgb_model.predict(last_x).reshape(pred_len, 1)

    # 각 row에 0을 붙여 avg_price만 복원
    dummy = np.zeros((pred_len, 2))
    pred_xgb_full = np.hstack([dummy, pred_xgb_scaled])
    pred_xgb = scaler.inverse_transform(pred_xgb_full)
    y_pred = pred_xgb[:, -1]

    rmse = mean_squared_error(y_true, y_pred, squared=False)
    mae = mean_absolute_error(y_true, y_pred)
    results_xgb.append((year, rmse, mae))

    # ---------- 시각화 ----------
    plt.figure(figsize=(10, 4))
    plt.plot(df_test['date'][:pred_len], y_true, label='실제 avg_price')
    plt.plot(df_test['date'][:pred_len], pred_lstm[:, -1], label='LSTM')
    plt.plot(df_test['date'][:pred_len], pred_xgb[:, -1], label='XGBoost')
    plt.title(f'{year}년 5월 평균가격 예측')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
print("\n📊 연도별 LSTM 성능:")
for y, r, m in results_lstm:
    print(f"{y}년 → RMSE: {r:.2f}, MAE: {m:.2f}")

print("\n📊 연도별 XGBoost 성능:")
for y, r, m in results_xgb:
    print(f"{y}년 → RMSE: {r:.2f}, MAE: {m:.2f}")
```
