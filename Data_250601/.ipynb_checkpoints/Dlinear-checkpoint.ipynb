```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl

# 한글 폰트 설정 (Mac의 경우)
mpl.rc('font', family='AppleGothic')
mpl.rcParams['axes.unicode_minus'] = False  # 마이너스 깨짐 방지
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from einops import rearrange
from sklearn.metrics import mean_squared_error, mean_absolute_error

# 1. 데이터 로딩 및 날짜 정렬
df = pd.read_csv("store/cabbage_separated.csv")
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
df = df.sort_values('date').reset_index(drop=True)

# 2. feature 선택 및 스케일링
features = ['intake', 'gap', 'avg_price']
df = df[features + ['date']].dropna()
train_df = df[df['date'] < '2025-05-01'].copy()
test_df  = df[(df['date'] >= '2025-05-01') & (df['date'] <= '2025-05-28')].copy()

scaler = StandardScaler()
scaled_data = scaler.fit_transform(train_df[features])

# 3. DLinear 정의
class DLinear(nn.Module):
    def __init__(self, seq_len, pred_len, input_dim):
        super(DLinear, self).__init__()
        self.linear = nn.Linear(seq_len, pred_len)

    def forward(self, x):
        x = rearrange(x, 'b t c -> b c t')
        x = self.linear(x)
        return rearrange(x, 'b c t -> b t c')

# 4. Dataset
class TimeSeriesDataset(Dataset):
    def __init__(self, data, seq_len, pred_len):
        self.data = data
        self.seq_len = seq_len
        self.pred_len = pred_len

    def __len__(self):
        return len(self.data) - self.seq_len - self.pred_len

    def __getitem__(self, idx):
        x = self.data[idx:idx + self.seq_len]
        y = self.data[idx + self.seq_len : idx + self.seq_len + self.pred_len]
        return torch.tensor(x, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

# 5. 학습 함수
def train_dlinear(model, dataloader, epochs=30, lr=0.001):
    model.train()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.MSELoss()
    for epoch in range(epochs):
        for x, y in dataloader:
            optimizer.zero_grad()
            loss = criterion(model(x), y)
            loss.backward()
            optimizer.step()

# 6. 학습 및 예측
seq_len = 30
pred_len = 28

# 학습용 dataloader
dataset = TimeSeriesDataset(scaled_data, seq_len, pred_len)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

model = DLinear(seq_len=seq_len, pred_len=pred_len, input_dim=3)
train_dlinear(model, dataloader)

# 예측 시점: 2025-04-30 직전 30일
target_index = train_df.shape[0]
input_seq = scaled_data[target_index - seq_len : target_index]
input_seq = torch.tensor(input_seq, dtype=torch.float32).unsqueeze(0)

# 예측 실행
model.eval()
with torch.no_grad():
    pred_scaled = model(input_seq).squeeze(0).numpy()
    pred = scaler.inverse_transform(pred_scaled)
    pred_price = pred[:, -1]  # avg_price만 추출

# 실제값
true_price = test_df['avg_price'].values[:pred_len]
dates = test_df['date'].values[:pred_len]

# 성능 출력
rmse = mean_squared_error(true_price, pred_price)
mae = mean_absolute_error(true_price, pred_price)
print(f"✅ DLinear (2025년 5월 평균가격 예측)")
print(f"RMSE: {rmse:.2f} / MAE: {mae:.2f}")

# 7. 시각화
plt.figure(figsize=(12, 6))
plt.plot(dates, true_price, label='실제 평균가격')
plt.plot(dates, pred_price, 'ro-', label='예측 평균가격')
plt.title("DLinear 배추 평균가격 예측 (2025년 5월)")
plt.xlabel("날짜")
plt.ylabel("가격 (원)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
