## 배추,HIGH - Dlinear로 batch학습

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from einops import rearrange
import matplotlib.pyplot as plt

# 1. DLinear 모델 정의
class DLinear(nn.Module):
    def __init__(self, seq_len, pred_len, input_dim):
        super(DLinear, self).__init__()
        self.linear = nn.Linear(seq_len, pred_len)

    def forward(self, x):
        x = rearrange(x, 'b t c -> b c t')  # [B, C, T]
        out = self.linear(x)                # [B, C, pred_len]
        out = rearrange(out, 'b c t -> b t c')
        return out

# 2. Dataset 클래스
class TimeSeriesDataset(Dataset):
    def __init__(self, data, seq_len, pred_len):
        self.data = data
        self.seq_len = seq_len
        self.pred_len = pred_len

    def __len__(self):
        return len(self.data) - self.seq_len - self.pred_len

    def __getitem__(self, idx):
        x = self.data[idx:idx+self.seq_len]
        y = self.data[idx+self.seq_len:idx+self.seq_len+self.pred_len]
        return torch.tensor(x, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

# 3. 데이터 로드 및 전처리 (HIGH만)
df = pd.read_csv("store/cabbage_separated.csv")
df = df[df['rate'] == 'HIGH']  # ✅ HIGH만 필터링
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
df = df.sort_values('date').reset_index(drop=True)

features = ['intake', 'gap', 'avg_price']
df = df[features + ['date']].dropna()

train_df = df[df['date'] < '2025-05-01'].copy()
test_df  = df[(df['date'] >= '2025-05-01') & (df['date'] <= '2025-05-28')].copy()

# 4. 스케일링
scaler = StandardScaler()
scaled_data = scaler.fit_transform(train_df[features])

# 5. 학습 준비
seq_len = 30
pred_len = 28
batch_size = 32
epochs = 30

dataset = TimeSeriesDataset(scaled_data, seq_len, pred_len)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

model = DLinear(seq_len=seq_len, pred_len=pred_len, input_dim=3)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.MSELoss()

# 6. 학습
model.train()
for epoch in range(epochs):
    for x, y in dataloader:
        optimizer.zero_grad()
        loss = criterion(model(x), y)
        loss.backward()
        optimizer.step()

# 7. 예측 (2025-04-30 기준)
model.eval()
with torch.no_grad():
    target_idx = train_df.shape[0]
    input_seq = scaled_data[target_idx - seq_len : target_idx]
    input_tensor = torch.tensor(input_seq, dtype=torch.float32).unsqueeze(0)  # [1, seq_len, 3]
    pred_scaled = model(input_tensor).squeeze(0).numpy()
    pred = scaler.inverse_transform(pred_scaled)
    pred_price = pred[:, -1]  # avg_price만

# 8. 실제값 추출
actual_len = min(len(test_df), pred_len)
true_price = test_df['avg_price'].values[:actual_len]
dates = test_df['date'].values[:actual_len]
pred_price = pred_price[:actual_len]

# 9. 성능 평가
rmse = mean_squared_error(true_price, pred_price)
mae = mean_absolute_error(true_price, pred_price)

print(f"\n✅ DLinear (2025년 5월 평균가격 예측 - HIGH만 대상)")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")

# 10. 시각화
plt.figure(figsize=(12, 6))
plt.plot(dates, true_price, label='실제 평균가격')
plt.plot(dates, pred_price, 'ro-', label='예측 평균가격')
plt.title("DLinear 배추 평균가격 예측 (2025년 5월, HIGH 등급)")
plt.xlabel("날짜")
plt.ylabel("가격 (원)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from einops import rearrange
import matplotlib.pyplot as plt

# 1. DLinear 정의
class DLinear(nn.Module):
    def __init__(self, seq_len, pred_len, input_dim):
        super(DLinear, self).__init__()
        self.linear = nn.Linear(seq_len, pred_len)

    def forward(self, x):
        x = rearrange(x, 'b t c -> b c t')     # [B, C, T]
        x = self.linear(x)                     # [B, C, pred_len]
        return rearrange(x, 'b c t -> b t c')  # [B, pred_len, C]

# 2. Dataset 정의
class TimeSeriesDataset(Dataset):
    def __init__(self, data, seq_len, pred_len):
        self.data = data
        self.seq_len = seq_len
        self.pred_len = pred_len

    def __len__(self):
        return len(self.data) - self.seq_len - self.pred_len

    def __getitem__(self, idx):
        x = self.data[idx:idx+self.seq_len]
        y = self.data[idx+self.seq_len:idx+self.seq_len+self.pred_len]
        return torch.tensor(x, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

# 3. 데이터 로드 + HIGH 필터 + lag 추가
df = pd.read_csv("store/cabbage_separated.csv")
df = df[df['rate'] == 'HIGH'].copy()
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
df = df.sort_values('date').reset_index(drop=True)

# lag feature 추가
df['avg_price_lag1'] = df['avg_price'].shift(1)
df['avg_price_lag7'] = df['avg_price'].shift(7)

features = ['intake', 'gap', 'avg_price', 'avg_price_lag1', 'avg_price_lag7']
df = df[features + ['date']].dropna()

train_df = df[df['date'] < '2025-05-01']
test_df  = df[(df['date'] >= '2025-05-01') & (df['date'] <= '2025-05-28')]

# 4. 스케일링
scaler = StandardScaler()
scaled_data = scaler.fit_transform(train_df[features])

# 5. 학습 설정
seq_len = 60
pred_len = 28
batch_size = 32
epochs = 30

dataset = TimeSeriesDataset(scaled_data, seq_len, pred_len)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

model = DLinear(seq_len=seq_len, pred_len=pred_len, input_dim=len(features))
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.L1Loss()

# 6. 학습
model.train()
for epoch in range(epochs):
    for x, y in dataloader:
        optimizer.zero_grad()
        loss = criterion(model(x), y)
        loss.backward()
        optimizer.step()
    if (epoch+1) % 10 == 0:
        print(f"[Epoch {epoch+1}] Loss: {loss.item():.4f}")

# 7. 예측 (2025-04-30 기준)
model.eval()
with torch.no_grad():
    last_input = scaled_data[-seq_len:]
    x_tensor = torch.tensor(last_input, dtype=torch.float32).unsqueeze(0)  # [1, seq_len, C]
    pred_scaled = model(x_tensor).squeeze(0).numpy()

    # 복원: avg_price만 정확히 inverse_transform
    last_row = train_df[features].values[-1]
    dummy = np.tile(last_row, (pred_len, 1))
    dummy[:, 2] = pred_scaled[:, 2]  # avg_price index = 2
    pred = scaler.inverse_transform(dummy)
    pred_price = pred[:, 2]

# 8. 실제값 비교
actual_len = min(len(test_df), pred_len)
true_price = test_df['avg_price'].values[:actual_len]
dates = test_df['date'].values[:actual_len]
pred_price = pred_price[:actual_len]

# 9. 평가 및 시각화
rmse = mean_squared_error(true_price, pred_price)
mae = mean_absolute_error(true_price, pred_price)

print(f"\n✅ 최적화된 DLinear 예측 결과 (HIGH 대상)")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")

plt.figure(figsize=(12, 6))
plt.plot(dates, true_price, label='실제 평균가격')
plt.plot(dates, pred_price, 'ro-', label='예측 평균가격')
plt.title("DLinear 예측 (2025년 5월, HIGH, lag + MAE 최적화)")
plt.xlabel("날짜")
plt.ylabel("가격 (원)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from einops import rearrange
import matplotlib.pyplot as plt

# 하이퍼파라미터
SEQ_LEN = 30
PRED_LEN = 28
BATCH_SIZE = 32
EPOCHS = 30
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# 1. 데이터 로드
df = pd.read_csv("store/cabbage_separated.csv")
df = df[df['rate'] == 'HIGH'].copy()
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
df = df.sort_values('date').reset_index(drop=True)

# 2. log 변환
df['avg_price'] = df['avg_price'].apply(lambda x: np.log1p(x))

# 3. lag feature
df['lag1'] = df['avg_price'].shift(1)
df['lag7'] = df['avg_price'].shift(7)
df = df.dropna().reset_index(drop=True)

# 4. Train/Test split
train_df = df[df['date'] < '2025-05-01'].copy()
test_df = df[(df['date'] >= '2025-05-01') & (df['date'] <= '2025-05-28')].copy()

# 5. 스케일링 (입력값만)
scaler = StandardScaler()
scaled_x = scaler.fit_transform(train_df[['intake', 'gap', 'lag1', 'lag7']])
target_y = train_df['avg_price'].values

# 6. Dataset 클래스
class TimeSeriesDataset(Dataset):
    def __init__(self, x, y, seq_len, pred_len):
        self.x = x
        self.y = y
        self.seq_len = seq_len
        self.pred_len = pred_len

    def __len__(self):
        return len(self.x) - self.seq_len - self.pred_len

    def __getitem__(self, idx):
        x_seq = self.x[idx:idx + self.seq_len]
        y_seq = self.y[idx + self.seq_len: idx + self.seq_len + self.pred_len]
        return torch.tensor(x_seq, dtype=torch.float32), torch.tensor(y_seq, dtype=torch.float32)

# 7. DLinear 모델 정의
class DLinear(nn.Module):
    def __init__(self, seq_len, pred_len, input_dim):
        super().__init__()
        self.linear = nn.Linear(seq_len, pred_len)

    def forward(self, x):
        x = rearrange(x, 'b t c -> b c t')         # [B, C, T]
        out = self.linear(x)                       # [B, C, pred_len]
        out = rearrange(out, 'b c t -> b t c')     # [B, pred_len, C]
        return out[:, :, -1]                       # avg_price만 사용

# 8. 데이터 준비
dataset = TimeSeriesDataset(scaled_x, target_y, SEQ_LEN, PRED_LEN)
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

model = DLinear(seq_len=SEQ_LEN, pred_len=PRED_LEN, input_dim=4).to(DEVICE)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# 9. 학습
model.train()
for epoch in range(1, EPOCHS + 1):
    total_loss = 0
    for x_batch, y_batch in loader:
        x_batch = x_batch.to(DEVICE)
        y_batch = y_batch.to(DEVICE)

        optimizer.zero_grad()
        output = model(x_batch)
        loss = criterion(output, y_batch)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"[Epoch {epoch}] Loss: {total_loss / len(loader):.4f}")

# 10. 예측
model.eval()
with torch.no_grad():
    latest_input = scaler.transform(train_df[['intake', 'gap', 'lag1', 'lag7']])[-SEQ_LEN:]
    latest_input_tensor = torch.tensor(latest_input, dtype=torch.float32).unsqueeze(0).to(DEVICE)
    pred = model(latest_input_tensor).cpu().squeeze(0).numpy()
    pred_price = np.expm1(pred)  # 복원

# 11. 실제값
true_price = np.expm1(test_df['avg_price'].values[:PRED_LEN])
dates = test_df['date'].values[:PRED_LEN]

# 12. 평가
rmse = mean_squared_error(true_price, pred_price)
mae = mean_absolute_error(true_price, pred_price)
print(f"\n✅ DLinear + log + lag 예측 (HIGH 대상)")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")

# 13. 시각화
plt.figure(figsize=(12, 6))
plt.plot(dates, true_price, label="실제값")
plt.plot(dates, pred_price, 'ro-', label="예측값")
plt.title("DLinear + lag + log 예측 (2025년 5월, HIGH)")
plt.xlabel("날짜")
plt.ylabel("평균 가격 (원)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
